```python
class image: 
    def __init__(a,image_path):
        import matplotlib.pyplot as plt
        import numpy as np
        from PIL import Image as im
        a.__plt=plt
        a.__np=np
        a.__im=im
        a.__image_path=image_path
        a.__image=im.open('{}'.format(a.__image_path)).convert('L')
        a.__image_aar=np.asarray(a._image__image)
    def display_imageO(a):
        image=a._image__im.fromarray(a._image__image_aar)
        return display(image)
    def histogram_for_image(a):
        a._image__plt.hist(a._image__image_aar.copy().reshape(-1))
        a._image__plt.show()      
    def increase_brightness(a,brightnesslevel=0):
        br=brightnesslevel
        l =a._image__image_aar.astype('int')+br
        new_arr=a._image__np.clip(l,0,255).astype('uint8')
        return new_arr
    def decrease_brightness(a,brightnesslevel=0):
        br=brightnesslevel
        t=a._image__image_aar.astype('int')-br
        new_arr1=a._image__np.clip(t,0,255).astype('uint8')
        return new_arr1
    def Histogram(a,x):
        new=x.reshape(-1)
        a._image__plt.hist(new)
        return a._image__plt.show()  
    def contrast(a,contrastlevel=0):
        cl=contrastlevel
        mean=a._image__image_aar.mean()
        data= a._image__image_aar.astype("int")
        new = ((data-mean)*cl+mean).astype('int')
        new=a._image__np.clip(new,0,255)
        new=new.astype("uint8")
        return new
    def noise_add(a,noiselevel=0):
        nl=noiselevel
        y=a._image__image_aar.copy()
        y=a._image__np.array(y)
        n=y.shape[0]*y.shape[1]
        u=a._image__np.random.normal(50,16,n)
        u=u.reshape(y.shape[0],y.shape[1])
        for i in range(y.shape[0]):
            for j in range(y.shape[1]):
                if u[i,j] < nl:
                    y[i,j] = 255
                elif u[i,j] > 100-nl:
                    y[i,j]=0
                else:
                    y[i,j]=y[i,j]
        return y
    def dispaly_image(a,x):
        new=x
        b=a._image__im.fromarray(new)
        return display(b)
    def median_filter(a,x):
        new=a._image__image_aar.copy()
        new2=a._image__image_aar.copy()
        for i in range(new.shape[0]-2):
            for j in range(new.shape[1]-2):
                new1 = new[i:i+3,j:j+3].astype('int')
                r = a._image__np.median(new1)
                new2[i+1,j+1] = r 
        return new2
```

```python
q=image(image_path='new.jpg')
```
